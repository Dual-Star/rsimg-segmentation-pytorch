```python
path_s1as_scene1 = 'data/s1-dset/s1as-scene/scene01_s1as.tif'
path_s1des_scene1 = 'data/s1-dset/s1des-scene/scene01_s1des.tif'
path_s1truth_scene1 = 'data/s1-dset/s1-truth/scene01_truth.tif'
path_s2_scene1 = 'data/s2-dset/s2-scene/scene01_20190125.tif'
path_s2_truth1 = 'data/s2-dset/s2-truth/scene01_20190125_truth.tif'

```
